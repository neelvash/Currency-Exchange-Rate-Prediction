Now we move onto the second part of cleaning the data, winsorization. 

This process will help to deal with outliers in the data. This was chosen over trimming as it reduces the influence of extreme values without completely removing them. This is especially relevant when looking at fincancial data, as spikes and dips in the market are common and expected.

```python
# Calculate thresholds based on 5%
lower_threshold = usd_jpy_df.quantile(0.05)
upper_threshold = usd_jpy_df.quantile(0.95)

print("Lower threshold:", lower_threshold)
print("Upper threshold:", upper_threshold)

winsorized_df = usd_jpy_df.clip(lower=lower_threshold, upper=upper_threshold)

usd_jpy_df.plot(label='Original')
winsorized_df.plot(label='Winsorized')
plt.legend()
plt.show()
```

## Checking stationarity of data (ADF test)

I started by chosing the upper and lower thresholds at 5% for each respective tail. I then visualised the original and windorized data, which can be seen above. Now, we will do use the Augmented Dickey-Fuller (ADF) test to statistically verify stationarity. If the p-value is greater than 0.05, the null hypothesis of non-stationarity is not rejected. This means the data is likely stationary, and I can proceed with building the ARIMA model.

```python
# Pass the winsorized_df to the adfuller test
adf_result = adfuller(winsorized_df)

# Print the test statistic and p-value
print("ADF Statistic:", adf_result[0])
print("p-value:", adf_result[1])
```

(https://www.analyticsvidhya.com/blog/2021/06/statistical-tests-to-check-stationarity-in-time-series-part-1/) - refresher on AD test.

To determine whether the data is stationary or not, we interpret the results of the Augmented Dickey-Fuller (ADF) test. In this test, the null hypothesis is that the data is non-stationary (it has a unit root), and the alternative hypothesis is that the data is stationary.

Since the p-value is greater than the significance levels of 0.05, we fail to reject the null hypothesis. Therefore, we do not have enough evidence to conclude that the data is stationary.

Thus, based on the ADF test results, it appears that the data is likely non-stationary.

## Differencing and ADF on differenced data

To deal with this, we will apply differencing. This involves subtracting the previous value from each data point, and then once again doing the ADF test on the differenced data. If differencing makes our data stationary, it will be included in the ARIMA model as the d value in ARIMA(p,d,q).

```python
# Apply one difference (d = 1)
differenced_df = winsorized_df.diff().dropna()  # Drop the first NaN value

# Check for stationarity again (optional)
adf_result_diff = adfuller(differenced_df)

differenced_df.plot()
plt.show()

print("ADF Statistic (differenced):", adf_result_diff[0])
print("p-value (differenced):", adf_result_diff[1])
```

As can be seen above, The ADF statistic after applying differencing is now significantly more negative and the p-value is much smaller than 0.05, indicating strong evidence against the presence of a unit root at the 5% significance level. This suggests that differencing has successfully made the data stationary.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
plot_acf(differenced_df, lags=40, ax=ax1)
plot_pacf(differenced_df, lags=40, ax=ax2)
plt.tight_layout()
plt.show()
```

## Deciding on ARIMA (p,d,q) values

Interpreting the ACF plot:

Since there's a significant spike at lag 1, this suggests the presence of an autoregressive term (AR) in the model.  I'll use p = 1 as a starting point.

Interpreting the PACF plot:

Another significant spike at lag 1 indicates a potential moving average term (MA), so i'll also use q = 1 as a starting point.

Additional notes:

The remaining spikes within the confidence interval bounds (between -0.25 and 0.25) are not statistically significant and can be ignored for initial model building.

Choosing the ARIMA order:

Based on the ACF and PACF plots, a starting point for the ARIMA order could be (1, 1, 1), where:

1 is the number of significant spikes in the ACF plot (for AR terms) - p value
1 is the degree of differencing determined from the ADF test - d value
1 is the number of significant spikes in the PACF plot (for MA terms) - q value

```python
import warnings
warnings.filterwarnings("ignore")  # To suppress ARIMA convergence warnings

window_size = 75
step_size = 1

online_forecasts = []
online_rmses = []

# Process data in a rolling window
for i in range(train_size, len(winsorized_df) - step_size, step_size):
    train_window = winsorized_df[i - window_size : i]
    test_window = winsorized_df[i : i + step_size]

    try:
        model = ARIMA(train_window, order=(15, 1, 1))  # Retrain model with each window
        model_fit = model.fit()

        forecast = model_fit.forecast(steps=step_size)[0]  # Forecast one step ahead
        online_forecast = model_fit.forecast(steps=step_size)  # For plotting

        inverted_forecast = pd.Series(forecast, index=test_window.index)
        for j in range(len(inverted_forecast)):
            inverted_forecast[j] += differenced_df[i + j - 1]  # Backshift to original scale

        # Combine forecasts
        online_forecasts.append(inverted_forecast)

        # Calculate and store metric
        online_rmse = mean_squared_error(test_window, inverted_forecast, squared=False)
        online_rmses.append(online_rmse)

    except:
        print("ARIMA fitting failed for window", i)
        online_forecasts.append(pd.Series(np.nan, index=test_window.index))
        online_rmses.append(np.nan)

# Combine online forecasts into a single Series
online_forecasts = pd.concat(online_forecasts)
```

```python
# Plot offline model, online model, and actual data

plt.figure(figsize=(12, 6))
plt.plot(winsorized_df, label='actual')
plt.plot(online_forecasts, label='online model')
plt.plot(forecast_results, label='offline model')

plt.legend()
plt.title("Offline vs Online ARIMA Predictions vs Actual")
plt.xlabel("Date")
plt.ylabel("USD-JPY Exchange Rate")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

# Print RMSE for online model
print("Online model RMSE:", np.mean(online_rmses))
```
